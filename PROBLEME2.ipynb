# CODE

```python
df=data.copy()
df.columns
```

```python
colonnes_a_supprimer = ["Année", "Code IRIS", "Nom IRIS", "Numéro de voie",
       "Indice de répétition", "Type de voie", "Libellé de voie",
       "Code Commune", "Nom Commune", "Segment de client",
       "Nombre de logements",
       "Consommation annuelle totale de l'adresse (MWh)", 'Consommation annuelle moyenne de la commune (MWh)', 'Adresse',
       'Code EPCI', 'Code Département', 'Code Région', 'Tri des adresses', 'N°_voie_(BAN)', 'Nom__rue_(BAN)',
       'Nom__commune_(BAN)', 'Type_énergie_générateur_ECS_n°2', 'Type_énergie_générateur_n°2_installation_n°2']
```

```python
df = df.drop(columns = colonnes_a_supprimer)
df.head()
```

```python
df = df[df['Type_énergie_n°1'].isin(['ELECTRICITE', 'ÉLECTRICITÉ'])]
df
```

```python
df = df[df['Type_bâtiment'].isin(['APPARTEMENT','MAISON'])]
df
```

```python
df['Consommation annuelle moyenne par logement de l\'adresse / m² (MWh/m²)'] = df['Consommation annuelle moyenne par logement de l\'adresse (MWh)'] / df['Surface_habitable_logement']
df['Conso_5_usages_é_finale_énergie_n°1/m²'] = df['Conso_5_usages_é_finale_énergie_n°1'] / df['Surface_habitable_logement']
df
```

```python
df['Ecart_pred'] = 1000*(-df['Consommation annuelle moyenne par logement de l\'adresse / m² (MWh/m²)'])+df['Conso_5_usages_é_finale_énergie_n°1/m²']
df['Ecart_pred_abs'] = df['Ecart_pred'].abs()
df['Ecart_pred_abs_pourcentage'] = (df['Ecart_pred_abs'] / (1000 * df["Consommation annuelle moyenne par logement de l'adresse / m² (MWh/m²)"])) * 100
df.head()
```

```python
print("Ecart minimum : %s" %df['Ecart_pred'].min())
print("Ecart maximum : %s" %df['Ecart_pred'].max())
print("Ecart moyen : %s" %df['Ecart_pred'].mean())
print("Premier quartile : %s" %df["Ecart_pred"].quantile(0.25))
print("Quantile 90 : %s" %df["Ecart_pred"].quantile(0.9))
print()

print("Ecart absolu minimum : %s" %df['Ecart_pred_abs'].min())
print("Ecart absolu maximum : %s" %df['Ecart_pred_abs'].max())
print("Ecart absolu moyen : %s" %df['Ecart_pred_abs'].mean())
print("Premier quartile : %s" %df["Ecart_pred_abs"].quantile(0.25))
print("Quantile 90 : %s" %df["Ecart_pred_abs"].quantile(0.9))
print()

print("Pourcentage d'écart minimum : %s" %df['Ecart_pred_abs_pourcentage'].min())
print("Pourcentage d'écart maximum : %s" %df['Ecart_pred_abs_pourcentage'].max())
print("Pourcentage d'écart moyen : %s" %df['Ecart_pred_abs_pourcentage'].mean())
print("Premier quartile : %s" %df["Ecart_pred_abs_pourcentage"].quantile(0.25))
print("Quantile 90 : %s" %df["Ecart_pred_abs_pourcentage"].quantile(0.9))
```

Supprimer les 10% de valeurs les plus élevées (abherrantes pour la plupart) ?

```python
df_90 = df[df["Ecart_pred_abs_pourcentage"] <= df["Ecart_pred_abs_pourcentage"].quantile(0.9)]

print("Ecart minimum : %s" %df_90['Ecart_pred'].min())
print("Ecart maximum : %s" %df_90['Ecart_pred'].max())
print("Ecart moyen : %s" %df_90['Ecart_pred'].mean())
print()

print("Ecart absolu minimum : %s" %df_90['Ecart_pred_abs'].min())
print("Ecart absolu maximum : %s" %df_90['Ecart_pred_abs'].max())
print("Ecart absolu moyen : %s" %df_90['Ecart_pred_abs'].mean())
print()

print("Pourcentage d'écart absolu minimum : %s" %df_90['Ecart_pred_abs_pourcentage'].min())
print("Pourcentage d'écart absolu maximum : %s" %df_90['Ecart_pred_abs_pourcentage'].max())
print("Pourcentage d'écart absolu moyen : %s" %df_90['Ecart_pred_abs_pourcentage'].mean())
```

# VISUALISATION

```python
erreur_appartement = df_90[df_90["Type_bâtiment"] == "APPARTEMENT"].groupby('Etiquette_DPE')['Ecart_pred'].mean()

erreur_maison = df_90[df_90["Type_bâtiment"] == "MAISON"].groupby('Etiquette_DPE')['Ecart_pred'].mean()

erreurs_combinees = pd.DataFrame({
    'APPARTEMENT': erreur_appartement,
    'MAISON': erreur_maison
})

erreurs_combinees.plot(kind='bar', title='Erreur moyenne par étiquette DPE et type de bâtiment', ylabel="Erreur d'estimation (kWh/m²)", xlabel='Étiquette DPE', rot=0)
plt.legend(title='Type de bâtiment')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
erreur_appartement = df_90[df_90["Type_bâtiment"] == "APPARTEMENT"].groupby('Etiquette_DPE')['Ecart_pred_abs_pourcentage'].mean()

erreur_maison = df_90[df_90["Type_bâtiment"] == "MAISON"].groupby('Etiquette_DPE')['Ecart_pred_abs_pourcentage'].mean()

erreurs_combinees = pd.DataFrame({'APPARTEMENT': erreur_appartement,'MAISON': erreur_maison})

erreurs_combinees.plot(kind='bar', title="Pourcentage d'erreur absolue moyenne par étiquette DPE et type de bâtiment", ylabel="Pourcentage d'erreur d'estimation absolue (%)", xlabel='Étiquette DPE', rot=0)
plt.legend(title='Type de bâtiment')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
filtres = ["Qualité_isolation_enveloppe", "Période_construction", "Type_énergie_principale_chauffage", "Type_installation_chauffage_n°1", "Type_bâtiment"]
for col in filtres:
    if col in df_90.columns :
        erreurs = df_90.groupby(col)['Ecart_pred'].mean()
        pourcentages = df_90.groupby(col)['Ecart_pred_abs_pourcentage'].mean()

        fig, axes = plt.subplots(1, 2, figsize=(16, 5))

        erreurs.sort_values().plot(kind='bar', ax=axes[0], title=f"Erreur moyenne par {col}")
        axes[0].set_ylabel("Erreur moyenne (kWh/m²)")
        axes[0].set_xlabel(col)
        axes[0].grid(axis='y', linestyle='--', alpha=0.7)
        axes[0].tick_params(axis='x', rotation=60)
        axes[0].set_xticklabels(axes[0].get_xticklabels(), ha='right', fontsize = 8)

        pourcentages.sort_values().plot(kind='bar', ax=axes[1], color='orange', title=f"Pourcentage d'erreur moyenne absolue par {col}")
        axes[1].set_ylabel("Pourcentage d'erreur moyenne absolue (%)")
        axes[1].set_xlabel(col)
        axes[1].grid(axis='x', linestyle='--', alpha=0.7)
        axes[1].tick_params(axis='x', rotation=60)
        axes[1].set_xticklabels(axes[1].get_xticklabels(), ha='right', fontsize = 8)

        plt.tight_layout()
        plt.grid()
        plt.show()
```

```python
filtres = ["Qualité_isolation_enveloppe", "Période_construction", "Type_énergie_principale_chauffage", "Type_installation_chauffage_n°1", "Type_bâtiment"]
for col in filtres:
    if col in df_90.columns:
        grouped = df_90.groupby(col)['Ecart_pred'].mean()
        grouped.sort_values().plot(kind='bar', figsize=(10, 5), title=f"Erreur moyenne par {col}")
        plt.ylabel("Erreur moyenne (kWh/m²)")
        plt.xlabel(col)
        plt.show()
```
